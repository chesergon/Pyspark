```python
from pyspark import SparkConf ,SparkContext
```

```python
# creating a SparkConf object
conf = SparkConf().setAppName("PysparkBasicsDemo").setMaster("local[*]")#"local[*]" runs spark locally using all cores 
# Create a sparkContext
sc = SparkContext(conf = conf)
```

```python
# Application Name
print("Application Name:", sc.appName)

# Master URL
print("Master URL:", sc.master)

# SparkConf - All configurations
print("SparkConf Settings:", sc.getConf().getAll())

# Default Parallelism (number of partitions)
print("Default Parallelism:", sc.defaultParallelism)

# Default Minimum Partitions for RDDs
print("Default Minimum Partitions:", sc.defaultMinPartitions)

# Spark Home Directory (if set)
print("Spark Home:", sc.sparkHome)

# User Running the Spark Application
print("Spark User:", sc.sparkUser())

# Spark and Python Versions
print("Spark Version:", sc.version)
print("Python Version:", sc.pythonVer)

# Registered RDDs (shows active RDDs)
print("Registered RDDs:", sc._jsc.getPersistentRDDs())

# Available Resources (CPU, Memory, etc.)
print("Available Resources:", sc._jsc.sc().resources())
```

```python
# Modify Log Level
sc.setLogLevel("WARN")  # Set log level to "WARN"

# Retrieve Current Log Level
print("Current Log Level:", sc.getConf().get("spark.logConf"))
```

```python
# Create a simple RDD from a range of numbers
rdd = sc.parallelize(range(10))

# Perform an action - sum the elements
sum_result = rdd.sum()
print("Sum of RDD Elements:", sum_result)
```

```python
# Stop the SparkContext
sc.stop()
```
